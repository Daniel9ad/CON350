```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv('../ciudadanos.csv', sep=';')
df
```

Eliminando filas que contengan nombres con numeros y ci con letras

```python
cn1 = ~(df['nombre1'].str.contains('\d', regex=True)==True) 
cn2 = ~(df['nombre2'].str.contains('\d', regex=True)==True)
cp = ~(df['paterno'].str.contains('\d', regex=True)==True)
cm = ~(df['materno'].str.contains('\d', regex=True)==True)
cci = ~(df['ci'].str.contains('[a-zA-Z]', regex=True)==True)
df = df[cn1 & cn2 & cp & cm & cci]
df
```

Uniendo nombres y apellidos

```python
def unir(nom1, nom2):
    if type(nom1) == float:
        return nom2
    elif type(nom2) == float:
        return nom1
    elif type(nom1) == float and type(nom2) == float:
        return ''
    return f'{nom1} {nom2}'
```

```python
df1 = df.copy()
df1['nombres'] = df.apply(lambda row: unir(row['nombre1'], row['nombre2']), axis=1)
df1['apellidos'] = df.apply(lambda row: unir(row['paterno'], row['materno']), axis=1)
df1
```

Convirtiendo el formato de fecha

```python
def convertir_fecha(fecha):
    partes = fecha.split('/')
    if len(partes) == 3:
        return f'{partes[2]}-{partes[1]}-{partes[0]}'
    else:
        return fecha
```

```python
df1 = df1.copy()
df1['fecha'] = df1['fecha'].apply(convertir_fecha)
df1
```

Guardando en la base de datos en este caso SQLITE

```python
conn = sqlite3.connect('../bd_personas.db')
```

```python
df1[['ci', 'nombres', 'apellidos', 'fecha']].to_sql('ciudadano', conn, if_exists='replace', index=False)
```

```python
conn.close()
```
